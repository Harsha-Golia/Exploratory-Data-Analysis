# EDA on books sales data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#data extraction -- csv data extracted to dataframe
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
#show first five rows
display(df.head())
```

```python
#view last 5 rows
display(df.tail())
```

```python
#display numerical statistics of the dataset - numerical summary of numerical columns
display(df.describe())
```

```python
 # Keep only books published in or after the year 1900
 # fetch gross sales which are greater than 1000

df = df[df['Publishing Year']>=1900]
display(df.head())
print(df.shape)



```

```python
#check how many null values in each columns
df.isna().sum()
```

```python
df.isnull().sum()
#drop the book name column which has null values
#handle the null values

```

```python
#handle null values

df.dropna(subset=['Book Name'], inplace=True)
df.fillna(0,inplace=True)

df.isnull().sum()
```

```python
#check count of duplicated rows in dataframe
print(df.duplicated().sum())
```

```python
#check unique entries
df.nunique()
```

```python
#create a histogram for the publishing year

df[df['gross sales']>1000]
print (df.shape)

book_name = df['Book Name']
publishing_year = df ['Publishing Year']
plt.hist(publishing_year, bins= 20, color='green', edgecolor= 'black')

#sns.histplot(df['Publishing Year'])
plt.title('Histogram of Publishing Year')
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.show()
```

```python
#genre fiction and other divisions
#categorical division of genre

df['genre'].value_counts()
```

```python
df['Author_Rating'].value_counts()
```

```python
#create a box plot to show book_ratings_count for each genre

df.boxplot(column='Book_ratings_count', by='genre')
plt.show()
```

```python
#create scatter plot for sale price and units sold

df.plot(kind='scatter', x='sale price', y='units sold')
plt.show()

#Insight - For lesser price there are more units sold and higher the sale price lesser will be the unit sold
```

```python
#view value counts for language code

df['language_code'].value_counts()
```

```python
#create a chart for language distribution

df2= df['language_code'].value_counts()
df2.plot(kind='barh')
plt.show()
```

```python
#group the data by publisher and calculate total revenue for each pubmisher

df.groupby('Publisher ')['publisher revenue'].sum().sort_values(ascending=False).reset_index().plot(kind='bar', x='Publisher ', y='publisher revenue')
plt.show()
```

```python
#average book rating count based on author rating

df.groupby('Author_Rating')['Book_ratings_count'].mean().reset_index()
```

```python
#next group by language code and group by number of books in each language

df.groupby('language_code')['Book Name'].count().reset_index()
```

```python
#Find maximum book ratings count based on author rating

max_ratingcount = df.groupby('Author_Rating')['Book_ratings_count'].max().sort_index()
max_ratingcount = max_ratingcount.reset_index()
display(max_ratingcount)
sns.lineplot(x = 'Author_Rating', y = 'Book_ratings_count', data= df)
plt.show()
```

```python
#PIE Chart : Share of Books by Author_Rating

import matplotlib.pyplot as plt
import matplotlib.cm as cm

Author_Rating_counts = df['Author_Rating'].value_counts().sort_index()
plt.figure(figsize=(6, 6))

plt.pie(Author_Rating_counts,
        labels=Author_Rating_counts.index,
        autopct='%1.1f%%',shadow= True,
        startangle=140,
        colors = cm.Pastel2.colors,
        textprops={'fontsize':10}
        )
plt.title('Book Distribution by by Author_Rating', fontsize = 14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot for Book_average_rating and Book_ratings_count

df.plot(kind='scatter', x='Book_average_rating', y='Book_ratings_count')
plt.xlabel('Book Average Rating')
plt.ylabel('Book Ratings Count')
plt.title('Book Rverage Ratings vs Book Ratings Count')
plt.show()

```

```python
# Create bar chart for total gross sales for each author

import matplotlib.pyplot as plt

total_gross_sales = df.groupby('Author')['gross sales'].sum().sort_values(ascending=False)
total_gross_sales.plot(kind='bar', color= 'blue')
plt.xlabel('Author')
plt.ylabel('Total Gross Sales')
plt.title('Total Gross Sales for Each Author')
plt.show()
```
